Student Performance Analysis

1.Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

2.Load Dataset

```python
df = pd.read_csv("student-mat.csv", sep=";")
print(df.shape)
df.head()
```

3.Explore and Clean Data

```python
# Check missing values
print(df.isnull().sum())
```

```python
# Remove duplicates
df = df.drop_duplicates()
print("Rows after removing duplicates:", len(df))
```

```python
# Data types
print(df.dtypes)
```

4.Analysis

```python
# Average final grade
print("Average G3:", df['G3'].mean().round(2))
```

```python
# Students who scored above 15
print("Students above 15:", (df['G3'] > 15).sum())
```

```python
# Correlation between study time and grade
print("Correlation (studytime vs G3):", df['studytime'].corr(df['G3']).round(4))
```

```python
# Average grade by gender
print(df.groupby('sex')['G3'].mean().round(2))
```

5.Visualizations

```python
# Histogram of final grades
plt.hist(df['G3'], bins=15, color='steelblue', edgecolor='black')
plt.title("Distribution of Final Grades")
plt.xlabel("Grade")
plt.ylabel("Number of Students")
plt.show()
```

```python
# Scatterplot: study time vs final grade
plt.scatter(df['studytime'], df['G3'], alpha=0.5)
plt.title("Study Time vs Final Grade")
plt.xlabel("Study Time (1=low, 4=high)")
plt.ylabel("Final Grade (G3)")
plt.show()
```

```python
# Bar chart: average grade by gender
gender_avg = df.groupby('sex')['G3'].mean()
gender_avg.plot(kind='bar', color=['salmon', 'steelblue'], edgecolor='black')
plt.title("Average Grade by Gender")
plt.xlabel("Gender")
plt.ylabel("Average G3")
plt.xticks(rotation=0)
plt.show()
```

6.Conclusions

Average final grade is 10.42 out of 20.
Only 40 students scored above 15.
Study time has a weak positive correlation with final grade (r = 0.098).
Male students score slightly higher on average than female students.
